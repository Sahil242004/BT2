```python
import fitz
text = ""
doc = fitz.open('resume.pdf')
for page in doc:
    text = text + page.get_text()
```

```python
print(text)
```

```python
def clean_text(text):
    return "\n".join(line.strip() for line in text.splitlines() if line.strip())

text = clean_text(text)
```

```python
print(text)
```
