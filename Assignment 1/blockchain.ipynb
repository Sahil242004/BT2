```python
import hashlib
```

```python
class Block:
  def __init__(self, data, prev_hash,nonce: int = 0):
    self.data = data
    self.prev_hash = prev_hash
    self.hash = self.calc_hash()
    self.nonce = nonce

  def calc_hash(self):
    sha = hashlib.sha256()
    sha.update(self.data.encode('utf-8'))
    return sha.hexdigest()
```

```python
class Blockchain:

  def __init__(self,difficulty=4):
    self.difficulty = difficulty
    self.chain = []
    self.create_genesis_block()

  
  def create_genesis_block(self):
        genesis = Block(data="genesis block", prev_hash="0")
        self.mine_block(genesis)
        self.chain.append(genesis)

  def last_block(self):
          return self.chain[-1]

  def is_valid_proof(self, block:Block) -> bool:
          return block.hash.startswith("0" * self.difficulty)

  def mine_block(self, block:Block) -> Block:
          while True:
              block.hash = block.calc_hash()
              if self.is_valid_proof(block):
                  return block
              block.nonce += 1
  
  def add_block(self, data):
    prev_block = self.chain[-1]
    new_block = Block(data, prev_block.hash)
    self.mine_block(new_block)
    if new_block.prev_hash != self.last_block().hash or not self.is_valid_proof(new_block):
            raise ValueError("Invalid block—rejected.")
    self.chain.append(new_block)

```

```python
blockchain = Blockchain()
```

```python
blockchain.add_block("This is the first block")
blockchain.add_block("This is the second block")
blockchain.add_block("This is the third block")
```

```python
print("Blockchain:")

for block in blockchain.chain:
    print('Data :',block.data)
    print('Privious Hash :', block.prev_hash)
    print('Hash :', block.hash)
    print()
```
